```python
import os
import glob2
from IPython.display import Image, display
import pandas as pd
import pipeline as pl  # import the functions from our pipeline
import sklearn.preprocessing
```

## Working with the project grant database

```python
projects = pl.load_and_peek_at_data('raw data/projects_2012_2013.csv',
                                 date_vars=['date_posted', 'datefullyfunded'],
                                 summary=False)
```

```python
# make dummy vars
cat_vars = ['primary_focus_subject', 'secondary_focus_area', 'secondary_focus_subject',
            'school_metro', 'school_charter', 'school_magnet',
           'primary_focus_area', 'resource_type', 'poverty_level', 'grade_level',
           'eligible_double_your_impact_match']
# make dummy outcome var
projects['60_day_outcome'] = (projects['datefullyfunded'] -
                              projects['date_posted']) > pd.Timedelta('60 days')
non_dummy_vars = ['date_posted', 'students_reached',
                'total_price_including_optional_support']
```

```python
projects = pl.make_dummies_subset_df(projects, cat_vars, non_dummy_vars)
```

```python
pl.fill_missing(projects, imputation_method='median')
```

```python
projects = pl.standardize_vars(projects, ['students_reached', 'total_price_including_optional_support'])
```

```python
cols_to_keep = [c for c in projects.columns if c.lower()[-2:] != '_f']
```

```python
projects = projects[cols_to_keep]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
kmeans.fit(projects)
```

```python
for cluster_label in np.unique(kmeans.labels_):
    indices_in_cluster = kmeans.labels_[np.where(kmeans.labels_ == cluster_label)]
    print('\n\n\n\n********************Cluster: ', cluster_label)
    description = projects.iloc[indices_in_cluster].describe()
    print('variables with 75th percentile >=1:')
    for col in description.columns:
        if description[col]['75%'] >= 1:
            print(col)
    description.to_csv('figures/summary_cluster_number' + str(cluster_label) + '.csv')
```

```python
means = []
for cluster_label in np.unique(kmeans.labels_):
    indices_in_cluster = kmeans.labels_[np.where(kmeans.labels_ == cluster_label)]
    current_cluster = projects.iloc[indices_in_cluster]
    for col in current_cluster.columns:
        current_mean = current_cluster[col].mean()
        if current_mean > 0:
            means.append((cluster_label, col, current_mean))  

means = pd.DataFrame(means)
means.columns = ['cluster', 'variable', 'mean']
means.head()
means['variable and cluster'] = means['variable'] + "        clstr: " + means['cluster'].astype(str)
```

```python
plt.figure(figsize=(16, 60))
sns.set(font_scale=1.5)
g = sns.barplot(y="variable and cluster", x="mean", data=means)
```

```python
means = []
for cluster_label in np.unique(kmeans.labels_):
    indices_in_cluster = kmeans.labels_[np.where(kmeans.labels_ == cluster_label)]
    current_cluster = projects.iloc[indices_in_cluster]
    for col in current_cluster.columns:
        current_mean = current_cluster[col].mean()
#         if current_mean > 0:
        means.append((cluster_label, col, current_mean))  

means = pd.DataFrame(means)
means.columns = ['cluster', 'variable', 'mean']
means.head()
means['variable and cluster'] = means['variable'] + "        clstr: " + means['cluster'].astype(str)
```

```python
means['mean'] = means['mean'] + (
    (np.random.uniform(size=means.shape[0])*2 - 1) / 5)   # adding some random jitter for plot
```

```python
df = means.pivot(index='variable', columns='cluster', values='mean').reset_index()
```

```python
df.head()
```

```python
# modified this code: https://python-graph-gallery.com/184-lollipop-plot-with-2-groups/
num_clusters = means['cluster'].unique().shape[0]

colors = ['red', 'blue', 'skyblue', 'hotpink', 'orange', 'purple',
          'green', 'yellow', 'black', 'grey', 'wheat', 'lavender', 'magenta',
          'darkblue', 'lime', 'forestgreen', 'navajowhite', 'olivedrab', 'cyan',
          'peru', 'azure', 'darkred']
mycolors = colors[:num_clusters]

# Reorder it following the values of the first value:
ordered_df = df
my_range=range(1,len(ordered_df.index)+1)


plt.figure(figsize=(16, 60))
# The vertical plot is made using the hline function
# I load the seaborn library only to benefit the nice looking feature
plt.hlines(y=my_range, xmin=means['mean'].min(), xmax=means['mean'].max(), color='grey', alpha=0.1)

for index, col in enumerate(ordered_df.columns[1:]):
    plt.scatter(ordered_df[col],
                my_range,
                color=mycolors[index],
                label=str(col), s=75, alpha=.5)
plt.legend()
 
# Add title and axis names
plt.yticks(my_range, ordered_df['variable'])
plt.title("Comparison of variable means by cluster, with random jitter added", loc='left')
plt.xlabel('Means')
plt.ylabel('Variable')
```
